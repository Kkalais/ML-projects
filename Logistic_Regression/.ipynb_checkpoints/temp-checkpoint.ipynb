# Problem 1: Gradient Descent

## a) expression for P(y|x)

**Data Assumption**:

\$\mathcal D = \{( x_1^{(i)}, x_2^{(i)}, y^{(i)})\},\ for \ i = 1,....,n  \$

**Output**

We want to estimate the weight vector $\mathbf w$.

**Model Assumption**: 

* Each response generated by the following model plus some Gaussian noise:
 \$y\_{i} = w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2} + \\epsilon\_i\$ 

* Noise $\epsilon$ is drawn from a Gaussian distribution

\$\$\\epsilon\_i \\sim N(0, \\sigma\^2)\$\$

* Each response y then becomes a draw from the following Gaussian:

\$\$\\Rightarrow y\_i|(\\mathbf{x}\_{1i} \\ ,\\mathbf{x}\_{2i}) \\sim
N(w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2}, \\sigma\^2) \$\$

* Probability of each response variable is
\$\$\\Rightarrow
P(y\_i|(\\mathbf{x}\_{1i} \\ ,\\mathbf{x}\_{2i}),\\mathbf{w})=\\frac{1}{\\sqrt{2\\pi\\sigma\^2}}e^{-\\frac{(y\_{i} - (w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2}))^{2}}{2\\sigma\^2}}\$\$ 

\$\$\\Rightarrow
P(y|(\\mathbf{x}\_{1} \\ ,\\mathbf{x}\_{2}),\\mathbf{w})=\prod_i\\frac{1}{\\sqrt{2\\pi\\sigma\^2}}e^{-\\frac{(y\_{i} - (w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2}))^{2}}{2\\sigma\^2}}\$\$ 

## b)Log likelihood function of training observations (x,y)
$$
 for \ \mathbf{x} = (x_{1}, x_{2} ) =>
$$

$$
\log \mathcal{L}(\omega|y,\mathbf{x}) = \log p(y|\mathbf{x},\omega) = \log \prod_i p(y_i|\mathbf{x_{i}},\omega) = \sum_i \log  p(y_i|\mathbf{x_{i}},\omega) = 
$$

\$\$\=
\sum_i \log \\frac{1}{\\sqrt{2\\pi\\sigma\^2}}e^{-\\frac{(y\_{i} - (w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2}))^{2}}{2\\sigma\^2}}\$\$

\$\$\=
\sum_i (- \frac{1}{2} \log (2\\pi\\sigma\^2) - \\frac{(y\_{i} - (w\_{0} + w\_{1} x\_{1i} + w\_{2} x\_{2i} + w\_{3} x\_{1i}^{2}))^{2}}{2\\sigma\^2} )\$\$

## c)Maximization of Log likelihood function 

* The desired parameter estimates: $$\omega = (\omega_{0},\omega_{1}, \omega_{2}, \omega_{3}) $$

$$
\omega =\mathop{\textrm{argmax}}_\omega(\log \mathcal{L}(\omega|y,\mathbf{x})) = 
$$
$$
\mathop{\textrm{argmax}}_\omega \sum_i (- \frac{1}{2} \log (2\pi\sigma^2) - \frac{(y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2}))^{2}}{2\sigma^2}) = 
$$
$$
\mathop{\textrm{argmax}}_\omega \sum_i ( - \frac{(y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2}))^{2}}{2\sigma^2})
$$

- We drop $\sigma^2$ because $\frac{1}{n}$ makes the loss interpretable

$$
\mathop{\textrm{argmin}}_\omega \sum_{i=0}^{n-1} \frac{(y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2}))^{2}}{n} \quad = \quad \mathop{\textrm{argmin}}_\omega (MSE)
$$
* So we can minimize the function :
$$
f(\omega = (\omega_{0},\omega_{1}, \omega_{2}, \omega_{3})) = \sum_{i=0}^{n-1} \frac{(y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2}))^{2}}{n}
$$

## d) Gradient of f(w)
$$
\nabla_{\omega} f(\omega) = \left[
\begin{array}{c}
\frac{\partial}{\partial \omega_0} f(\omega) \\
\frac{\partial}{\partial \omega_1} f(\omega) \\
\frac{\partial}{\partial \omega_2} f(\omega) \\
\frac{\partial}{\partial \omega_3} f(\omega)
\end{array}\right] =>
$$

$$
\nabla_{\omega} f(\omega) = \left[
\begin{array}{c}
(\frac{2}{n}) \sum_{i=0}^{n-1}[  (y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2})) \cdot (-1)  ]\\
(\frac{2}{n}) \sum_{i=0}^{n-1}[  (y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2})) \cdot (-x_{1i}) ] \\
(\frac{2}{n}) \sum_{i=0}^{n-1}[  (y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2})) \cdot (-x_{2i}) ] \\
(\frac{2}{n}) \sum_{i=0}^{n-1}[  (y_{i} - (w_{0} + w_{1} x_{1i} + w_{2} x_{2i} + w_{3} x_{1i}^{2})) \cdot (-x_{1i}^{2}) ]
\end{array}\right]
$$

## e) Gradient descent update rule for w in terms of ∇f(w)

- Evaluate gradients at each data point in the dataset, with the form above (d)
- Update parameters of the function to approximate the data 

$$\omega_{i} = \omega_{i}  - \alpha \frac{\partial}{\partial \omega_{i}} f(\omega) \\$$
$$for \ each \ i \ \epsilon \ (0,1,2,3), \ where \ a = learning \ rate$$

- \$\mathcal D = \{( x_1^{(i)}, x_2^{(i)}, y^{(i)})\},\ for \ i = 1,....,n  => \$


- \$\mathcal X = \{[ x_1^{(i)}, x_2^{(i)}]\}\ for \ i = 1,....,n \$


```python
from sympy import *
import pandas as pd
import numpy as np

# x=[x1|x2] ->independent variable's array
# y=[y(i)] for i=0,1,...,n-1 ->dependent variable

data=pd.read_csv('Gradient_descent.csv')
x = data.iloc[:,0:2].values
y = data.iloc[:,2:3].values

def update_weights(x, y, weight, learning_rate):
    weight_deriv = np.zeros(len(weight))
    
    for i in range(len(y)):
        temp = weight[0] + weight[1]*x[i][0] + weight[2]*x[i][1] + weight[3]*(x[i][0]**2)
        
        weight_deriv[0] = (-1) * (y[i] - temp)
        weight_deriv[1] = (-x[i][0]) * (y[i] - temp)
        weight_deriv[2] = (-x[i][1]) * (y[i] - temp)
        weight_deriv[3] = (-x[i][0]**2) * (y[i] - temp)

    weight -= (2*weight_deriv / len(y)) * learning_rate

    return weight
```

## f) We will compute the above derivatives using sympy python library

```python
w0 = Symbol('w0')
w1 = Symbol('w1')
w2 = Symbol('w2')
w3 = Symbol('w3')

#define function f=f(w)=f(w0,w1,w2,w3)
sum=0
for i in range(len(y)):
    sum += (y[i]- (w0 + w1*x[i][0] + w2*x[i][1] + w3*(x[i][0]**2)) )**2

f_w = sum/len(y)

#partial derivatives of f on w_i
df_dw0 = diff(f_w,w0)
df_dw1 = diff(f_w,w1)
df_dw2 = diff(f_w,w2)
df_dw3 = diff(f_w,w3)

grad_f_w = np.array([df_dw0,df_dw1,df_dw2,df_dw3])
```

# Problem 2: Logistic Regression

## a) Formula for the log likelihood of the training data

**Data Assumption**:

- \$\mathcal D = \{( x_1^{(i)}, x_2^{(i)},\dots ,x_n^{(i)},y^{(j)})\},\ for \ i = 1(1)p \ and \ j = 1(1)n \\ \$

- \$ eg. \mathbf{x_1}=[x_{11} \ x_{12} \ ..... \ x_{1p}] -> a \ row \ instance \ with \ p-features \$

**Output**

We want to estimate the weight vector $\mathbf w$.  

**Model Assumption**: 

* Each response generated by the following model plus some Gaussian noise:
 \$y\_{i} = w\_{0} + \mathbf{x_i} w + \\epsilon\_i\$ 

* Noise $\epsilon$ is drawn from a Gaussian distribution

\$\$\\epsilon\_i \\sim N(0, \\sigma\^2)\$\$

* Each response y then becomes a draw from the following Gaussian:

\$\$\\Rightarrow y\_i|\\mathbf{x}\_{i} \\sim
N(w\_{0} + \mathbf{x_i} w,\ \\sigma\^2) \$\$

* Probability of each response variable is
In logistic regression we model a binary variable $y_i \in \{0,1\}$

$$
\begin{aligned}
p(y_i=+1|\mathbf{x_i},w) &= σ(w_0 + \mathbf{x_i}w)\\
p(y_i=0|\mathbf{x_i},w) &= 1 - σ(w_0 + \mathbf{x_i}w) = σ(-(w_0 + \mathbf{x_i}w))
\end{aligned}
$$

This is equivalent to

$$
\begin{aligned}
p(y_i|\mathbf{x_i},w) = p(y_i=1|\mathbf{x_i},w)^{y_i} \cdot p(y_i=0|\mathbf{x_i},w)^{1-y_i} = {\frac{1}{1 + e^{-(w_0 + \mathbf{x_i}w)}}}^{y_i} \cdot {\frac{1}{1 + e^{+(w_0 + \mathbf{x_i}w)}}}^{1-y_i} =>
\end{aligned}
$$

$$
\begin{aligned}
p(y|x,w) &= \prod_{i=0}^{n-1}(\frac{1}{1 + e^{-(w_0 + \mathbf{x_i}w)}}^{y_i} \cdot {\frac{1}{1 + e^{+(w_0 + \mathbf{x_i}w)}}}^{1-y_i})
\end{aligned}
$$

Likelihood function is:
$$
\mathcal{L}(w|y,x) = \prod_{i=0}^{n-1}(\frac{1}{1 + e^{-(w_0 + \mathbf{x_i}w)}}^{y_i} \cdot {\frac{1}{1 + e^{+(w_0 + \mathbf{x_i}w)}}}^{1-y_i})
$$

Log-likelihood function is:
$$
f(w) = \log \mathcal{L}(w|y,x) = \sum_{i=0}^{n-1}[y_i \cdot \log(\frac{1}{1 + e^{-(w_0 + \mathbf{x_i}w)}}) + (1-y_i) \cdot \log(\frac{1}{1 + e^{+(w_0 + \mathbf{x_i}w)}})] = \\ 
$$

$$
= \sum_{i=0}^{n-1}[y_i \cdot \log(\frac{e^{+(w_0 + \mathbf{x_i}w)}}{1 + e^{+(w_0 + \mathbf{x_i}w)}}) + (y_{i}-1) \cdot \log(1 + e^{+(w_0 + \mathbf{x_i}w)})] = \\
$$

$$
= \sum_{i=0}^{n-1}[ y_i \cdot(w_0 + \mathbf{x_i}w) - y_i \cdot \log(1 + e^{+(w_0 + \mathbf{x_i}w)}) + y_{i} \cdot \log(1 + e^{+(w_0 + \mathbf{x_i}w)}) - \log(1 + e^{+(w_0 + \mathbf{x_i}w)}) ] = \\
$$

$$
= \sum_{i=0}^{n-1}[ y_i \cdot(w_0 + \mathbf{x_i}w) - \log(1 + e^{+(w_0 + \mathbf{x_i}w)}) ]
$$

## b) Partial derivatives of log-likelihood function
$$
\frac{\partial}{\partial \omega_0} f(\omega) = \sum_{0}^{n-1}[ y_i - \frac{e^{+(w_0 + \mathbf{x_i}w)}}{1 + e^{+(w_0 + \mathbf{x_i}w)}}]\\\\
\frac{\partial}{\partial \omega_j} f(\omega) = \sum_{0}^{n-1}[ y_{i}x_{ij} - \frac{e^{+(w_0 + \mathbf{x_i}w)}\cdot x_{ij}}{1 + e^{+(w_0 + \mathbf{x_i}w)}} ] , \ for \ each \ j \in (1, \dots , p) \\ 
$$

## c) Logistic regression classifier using gradient ascent

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import *
import numpy as np
import random

# Calculate the logistic regression objective value
def LR_CalcObj(XTrain,yTrain,wHat):

    # get dimensions
    n,p = XTrain.shape    
  
    obj=0
    # calculate objective value
    for i in range(n):
        temp = np.dot(XTrain[i,:].T,wHat)
        obj += (yTrain[i]*temp) - log(1+exp(temp))
           
    return obj

# Calculate the gradient of the logistic regression
# objective function with respect to each parameter
def LR_CalcGrad(XTrain,yTrain,wHat):

    # get dimensions
    n,p = XTrain.shape
    
    grad = np.ones(shape=(p,1))
    
    for j in range(p):
        for i in range(n):
            temp = np.dot(XTrain[i,:].T,wHat)
            temp = exp(temp)
            
            if (i!=0):
                grad[j] += (yTrain[i]*XTrain[i,j]) - (temp*XTrain[i,j])/(1+temp)
            else:
                grad[j] += yTrain[i] - temp/(1+temp)
     
    return(grad)
    
# Calculate the new value of wHat using the gradient 
# ascent update rule
def LR_UpdateParams(wHat,grad,eta):

    # update value of w
    wHat = wHat + eta*grad;

    return wHat

# Check whether the objective value has converged 
# by comparing the difference between consecutive 
# objective values to the tolerance
def LR_CheckConvg(oldObj,newObj,tol):

    # compute difference between objectives
    diff = abs(oldObj-newObj)
  
    # compare difference to tolerance
    if (diff < tol):
        hasConverged = True
    else:
        hasConverged = False
  
    return(hasConverged)
    
def LR_GradientAscent(XTrain,yTrain):

    # Set the step size
    eta = 0.01
    
    # Set the convergence tolerance
    tol = 0.001
    
    # Get dimensions
    n,p = XTrain.shape
    
    # Initialize wHat
    wHat = np.zeros((p,1))
    
    # Initialize objVals
    objvals = []
    objVals = np.append(objvals,LR_CalcObj(XTrain,yTrain,wHat))
    
    # Initialize convergence flag
    hasConverged = False
    
    # Run gradient ascent until convergence
    while (not hasConverged):
        # Calculate gradient
        grad = LR_CalcGrad(XTrain,yTrain,wHat)

        # Update parameter estimate
        wHat = LR_UpdateParams(wHat,grad,eta)

        # Calculate new objective
        newObj = LR_CalcObj(XTrain,yTrain,wHat)

        # Check convergence
        hasConverged = LR_CheckConvg(objVals[-1],newObj,tol)

        #Store new objective
        objVals = np.append(objVals,newObj)   

    return (wHat,objVals)

# Predict the labels for a test set using logistic regression
def LR_PredictLabels(XTest,yTest,wHat):

    # get dimensions
    m,p = XTest.shape
  
    # precompute X*w and exp(X*w)
    Xw = np.dot(XTest,wHat)
    eXw = np.exp(Xw) 
  
    # calculate p(Y=0)
    pY0 = 1/(1 + eXw)
    
    # calculate p(Y=1)
    pY1 = eXw/(1 + eXw)
  
    # choose best label
    max_pY=np.zeros((len(pY0),2))
    max_pY[:,0] = pY0[:,0]
    max_pY[:,1] = pY1[:,0]

    yHat = np.argmax(max_pY,axis=1)
    
    # calculate error
    numErrors=0
    for i in range(len(yHat)):
        if(yHat[i] != yTest[i]):
            numErrors += 1

    return (yHat,numErrors)
```

## d), e), f) Train classifier and predict labels

```python
def RunLR1():
    
    XTest=pd.read_csv('XTest.csv').values
    yTest=pd.read_csv('YTest.csv').values
    XTrain=pd.read_csv('XTrain.csv').values
    yTrain=pd.read_csv('YTrain.csv').values

    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTrain),1))
    XTrain = np.append(bias, XTrain, axis=1)
    
    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTest),1))
    XTest = np.append(bias, XTest, axis=1)
    
    # Train logistic regression
    wHat,objVals = LR_GradientAscent(XTrain,yTrain)

    # Test logistic regression
    yHat,numErrors = LR_PredictLabels(XTest,yTest,wHat)

    # Print the number of misclassified examples
    print('There were %d misclassified examples in the test set\n' %numErrors)

    # Plot the objective value per iteration
    plt.figure()
    plt.xlabel('Gradient Ascent Iteration')
    plt.ylabel('Logistic Regression Objective Value')
    plt.title('Convergence of Gradient Ascent for Logistic Regression')
    plt.plot(objVals)
    plt.show()

    print('Gradient ascent coverges after %d iterations\n' %(len(objVals)-1))
    
    return
```

```python
RunLR1()
```

## g) How the training and test error change as the training set size increases ?

```python
def RunLR2():   
    
    XTest=pd.read_csv('XTest.csv').values
    yTest=pd.read_csv('YTest.csv').values
    XTrain=pd.read_csv('XTrain.csv').values
    yTrain=pd.read_csv('YTrain.csv').values

    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTrain),1))
    XTrain = np.append(bias, XTrain, axis=1)
    
    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTest),1))
    XTest = np.append(bias, XTest, axis=1)
    
    # Evaluate the training and test error as a function of training set size
    n = len(XTrain)
    kVals = np.arange(10,len(XTrain),10)
    m = len(XTest)
    
    trainingError = np.zeros((len(kVals),1))
    testError = np.zeros((len(kVals),1))

    for kInd in range(len(kVals)):
        # Set k
        k = kVals[kInd]
        # Generate training set
        subsetInds=[]
        for i in range(k):
            temp=random.randint(1,n-1)
            if( temp in subsetInds):
                while(temp in subsetInds):
                    temp=random.randint(1,n-1)
            subsetInds=np.append(subsetInds,temp)
        
        XTrainSubset = np.zeros((k,len(XTrain[0])))
        for o in range(k):
            XTrainSubset[o]=XTrain[int(subsetInds[o]),:]
                
        yTrainSubset = np.zeros((k,len(yTrain[0])))
        for o in range(k):
            yTrainSubset[o]=yTrain[int(subsetInds[o])]
            
        # Train logistic regression
        wHat,objvals = LR_GradientAscent(XTrainSubset,yTrainSubset)
        # Test classifier on training set
        [yHatTrain,numErrorsTrain] = LR_PredictLabels(XTrainSubset,yTrainSubset,wHat)
        trainingError[kInd] = numErrorsTrain/k
        # Test classifier on test set
        [yHatTest,numErrorsTest] = LR_PredictLabels(XTest,yTest,wHat)
        testError[kInd] = numErrorsTest/m   
      
    plt.figure()
    plt.plot(kVals,trainingError,'b')
    plt.plot(kVals,testError,'r')
    plt.xlabel('Training Set Size')
    plt.ylabel('Prediction Error')
    plt.legend(['Training Error','Test Error'])
    plt.title('Logistic Regression Performance by Training Set Size')
    plt.show()
    
    return
```

```python
RunLR2()
```

```python
# Perform the same experiment but average over multiple random training sets

def RunLR3():
    
    XTest=pd.read_csv('XTest.csv').values
    yTest=pd.read_csv('YTest.csv').values
    XTrain=pd.read_csv('XTrain.csv').values
    yTrain=pd.read_csv('YTrain.csv').values

    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTrain),1))
    XTrain = np.append(bias, XTrain, axis=1)
    
    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTest),1))
    XTest = np.append(bias, XTest, axis=1)
    
    n = len(XTrain)
    kVals = np.arange(10,len(XTrain),10)
    m = len(XTest)
    
    trainingError = np.zeros((len(kVals),1))
    testError = np.zeros((len(kVals),1))
    
    for kInd in range(len(kVals)):
        # Set k
        k = kVals[kInd]
        
        for rep in range(2):
            # Generate training set
            subsetInds=np.zeros((k,1))
        
            for i in range(len(subsetInds)):
                temp=random.randint(1,n-1)
                if( temp in subsetInds):
                    while(temp in subsetInds):
                        temp=random.randint(1,n-1)
                subsetInds[i]=temp
                
            XTrainSubset = np.zeros((k,len(XTrain[0])))
            for o in range(k):
                XTrainSubset[o]=XTrain[int(subsetInds[o]),:]
                
            yTrainSubset = np.zeros((k,len(yTrain[0])))
            for o in range(k):
                yTrainSubset[o]=yTrain[int(subsetInds[o])]
            # Train logistic regression
            wHat,objVals = LR_GradientAscent(XTrainSubset,yTrainSubset)
            # Test classifier on training set
            [yHatTrain,numErrorsTrain] = LR_PredictLabels(XTrainSubset,yTrainSubset,wHat)
            trainingError[kInd] += numErrorsTrain/k
            # Test classifier on test set
            [yHatTest,numErrorsTest] = LR_PredictLabels(XTest,yTest,wHat)
            testError[kInd] += numErrorsTest/m
           
        trainingError[kInd] /= 100
        testError[kInd] /= 100
          
    plt.figure()
    plt.plot(kVals,trainingError,'b')
    plt.plot(kVals,testError,'r')
    plt.xlabel('Training Set Size')
    plt.ylabel('Average Prediction Error')
    plt.legend(['Training Error','Test Error'])
    plt.title('Average Logistic Regression Performance by Training Set Size')

    return
```

```python
RunLR3()
```

- With small size of training set, we have model overfitting. As the training set size increases, we avoid this because our model can't remember all training data, but it will try to find some general setting that applies to majority of training data for reducing loss during training. But same general settings can also be applicable to predict on test data. So by increasing training data, training loss is increased but test loss decreases. Also, we see that if the errors reach that asymptote, we can not improve the accuracy score by using more training data.

## h) Analytical expression for the decision boundary 

- The decision boundary is found by solving the equation:

$$
\begin{aligned}
p(y=1|\mathbf{x},w) = p(y=0|\mathbf{x},w) =>\\\\
σ(w_0 + \mathbf{x}w) = σ(-(w_0 + \mathbf{x}w)) =>\\\\
\frac{1}{1+e^{-(w_0 + \mathbf{x}w)}} = \frac{1}{1+e^{w_0 + \mathbf{x}w}} =>\\\\
\frac{e^{w_0 + \mathbf{x}w}}{1+e^{w_0 + \mathbf{x}w}} = \frac{1}{1+e^{w_0 + \mathbf{x}w}} =>\\\\
e^{w_0 + \mathbf{x}w} = 1 => w_0 + \mathbf{x}w = 0 =>\\\\
w_0 + w_1x_1 + w_2x_2 + \cdots + w_px_p = 0 \\\\
\end{aligned}
$$

- So, we found a linear expression for the decision boundary.
- i.e. considering only dimensions j and k, that have highest absolute weight in your estimated parameters w =>
$$
\begin{aligned}
\\w_0 + w_jx_j + w_kx_k = 0 => x_k = (\frac{-w_j}{w_k})\cdot x_j + (\frac{-w_0}{w_k})\\\\
\end{aligned}
$$

- So, if we know the values of parameters w, we don't need any further information to describe the decision boundary.

## i) Plotting the decision boundary 

```python
def PlotDB():

    XTest=pd.read_csv('XTest.csv').values
    yTest=pd.read_csv('YTest.csv').values
    XTrain=pd.read_csv('XTrain.csv').values
    yTrain=pd.read_csv('YTrain.csv').values

    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTrain),1))
    XTrain = np.append(bias, XTrain, axis=1)
    
    # add a feature of 1's to XTrain
    bias = np.ones(shape=(len(XTest),1))
    XTest = np.append(bias, XTest, axis=1)
    
    # Train logistic regression
    wHat,objVals = LR_GradientAscent(XTrain,yTrain)

    # Plot data in 2d
    plt.figure()
    j = 1
    k = 2
    ind0 = (yTest == 0)
    ind1 = (yTest == 1)
    
    temp1=XTest[:,j]
    temp1=np.reshape(temp1,(len(temp1),1))
    temp1=temp1[ind0]
    
    temp2=XTest[:,k]
    temp2=np.reshape(temp2,(len(temp2),1))
    temp2=temp2[ind0]
    
    temp3=XTest[:,j]
    temp3=np.reshape(temp3,(len(temp3),1))
    temp3=temp3[ind1]
    
    temp4=XTest[:,k]
    temp4=np.reshape(temp4,(len(temp4),1))
    temp4=temp4[ind1]
    
    plt.plot(temp1,temp2,'r.')
    plt.plot(temp3,temp4,'b.')
    
    # Calculate decision boundary
    dbDimJ = np.arange(min(XTest[:,j]),max(XTest[:,j]),0.01)
    dbDimK = -(wHat[0] + wHat[j]*dbDimJ)/wHat[k]
  
    # Plot decision boundary
    plt.plot(dbDimJ,dbDimK,'k-')
  
    # Set plot title and axis labels
    plt.xlabel('Dimension %d'%j)
    plt.ylabel('Dimension %d'%k)
    plt.title('Logistic Regression Decision Boundary')
    plt.show()

    return
```

```python
PlotDB()
```
